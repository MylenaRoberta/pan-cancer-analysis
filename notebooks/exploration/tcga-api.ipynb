# **Exploração da API do The Cancer Genome Atlas (TCGA)**
Desenvolvida pela Genomic Data Commons (GDC)

# Importação de Bibliotecas

```python
import json
import pandas as pd
import requests
```

# Constantes

```python
# URL HTML base da API TCGA / GDC
TCGA_API_ENDPOINT = 'https://api.gdc.cancer.gov'

ANNOTATIONS_ENDPOINT = f'{TCGA_API_ENDPOINT}/annotations'
CASES_ENDPOINT = f'{TCGA_API_ENDPOINT}/cases'
DATA_ENDPOINT = f'{TCGA_API_ENDPOINT}/data'
FILES_ENDPOINT = f'{TCGA_API_ENDPOINT}/files'
MANIFEST_ENDPOINT = f'{TCGA_API_ENDPOINT}/manifest'
PROJECTS_ENDPOINT = f'{TCGA_API_ENDPOINT}/projects'
SLICING_ENDPOINT = f'{TCGA_API_ENDPOINT}/slicing'
STATUS_ENDPOINT = f'{TCGA_API_ENDPOINT}/status'
SUBMISSION_ENDPOINT = f'{TCGA_API_ENDPOINT}/submission'
```

# Exploração Inicial dos Endpoints

## `status`

```python
# Verifica o status e a versão da API
response = requests.get(STATUS_ENDPOINT)

# Printa o conteúdo da resposta
print(response.content.decode('utf-8'))
```

## `projects`

```python
# Mapeia todas as informações contidas no endpoint
response = requests.get(f'{PROJECTS_ENDPOINT}/_mapping')

# Converte o conteúdo da resposta para JSON e a printa
json_response = json.loads(response.content.decode('utf-8'))
json_response_str = json.dumps(json_response, indent=2)
print(json_response_str)
```

```python
# Campos disponíveis no endpoint
fields = [
    'dbgap_accession_number',
    'disease_type',
    'intended_release_date',
    'name',
    'primary_site',
    'program.dbgap_accession_number',
    'program.name',
    'program.program_id',
    'project_autocomplete',
    'project_id',
    'releasable',
    'released',
    'state',
    'summary.case_count',
    'summary.data_categories.case_count',
    'summary.data_categories.data_category',
    'summary.data_categories.file_count',
    'summary.experimental_strategies.case_count',
    'summary.experimental_strategies.experimental_strategy',
    'summary.experimental_strategies.file_count',
    'summary.file_count',
    'summary.file_size'
]
fields = ','.join(fields)

# Busca por todos os projetos da API
response = requests.post(
    url=PROJECTS_ENDPOINT,
    headers={'Content-Type': 'application/json'},
    json={'fields': fields}
)

# Converte o conteúdo da resposta para JSON
json_response = json.loads(response.content.decode('utf-8'))['data']['hits']

# Printa um dos objetos do conteúdo da resposta
json_response_str = json.dumps(json_response[1], indent=2)
print(json_response_str)
```

## `cases`

```python
# Mapeia todas as informações contidas no endpoint
response = requests.get(f'{CASES_ENDPOINT}/_mapping')

# Converte o conteúdo da resposta para JSON e a printa
json_response = json.loads(response.content.decode('utf-8'))
json_response_str = json.dumps(json_response, indent=2)
print(json_response_str)
```

```python
# Campos de interesse no endpoint
fields = [
    'case_id',
    'files.access',
    'files.data_category',
    'files.data_format',
    'files.data_type',
    'files.experimental_strategy',
    'files.file_id',
    'files.file_name',
    'files.file_size',
    'project.project_id',
    'state'
]
fields = ','.join(fields)

# Parâmetros para a requisição ao endpoint
params = {
    'fields': fields,
    'format': 'json'
}

# Busca por todos os casos da API
response = requests.post(
    url=CASES_ENDPOINT,
    headers={'Content-Type': 'application/json'},
    json={'fields': fields, 'size': '1'}
)

# Converte o conteúdo da resposta para JSON
json_response = json.loads(response.content.decode('utf-8'))['data']['hits']

# Printa um dos objetos do conteúdo da resposta
json_response_str = json.dumps(json_response, indent=2)
print(json_response_str)
```
