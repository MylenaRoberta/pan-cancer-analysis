# Importação de Bibliotecas

```python
import numpy as np
import pandas as pd
```

# Constantes

```python
# URL HTML base da API GDC
GDC_API_ENDPOINT = 'https://api.gdc.cancer.gov'

# Endpoint para download dos arquivos
DATA_ENDPOINT = f'{GDC_API_ENDPOINT}/data'

# Caminho da pasta de dados externos
EXTERNAL_DATA_PATH = '../../data/external/prototyping'

# Caminho da pasta de dados intermediários
INTERIM_DATA_PATH = '../../data/interim/prototyping'
```

# Carregamento de Dados

```python
# DataFrame com os casos de interesse
df_cases = pd.read_csv(
    f'{INTERIM_DATA_PATH}/tcga-cases-of-interest.csv'
)

# DataFrame com os arquivos de interesse
df_files = pd.read_csv(
    f'{INTERIM_DATA_PATH}/tcga-files-of-interest.csv'
)

# Cruzamento entre os DataFrames de casos e arquivos
df_eda = df_cases \
    .merge(
        right=df_files,
        left_on='case_id',
        right_on='case_id',
        how='inner'
    )

# 
df_eda['sample_acronym'] = np.where(
    df_eda['tissue_type'] == 'Normal', 'NT', 'TT'
)

df_eda['data_acronym'] = np.where(
    df_eda['data_type'] == 'Gene Expression Quantification', 'mRNA', 'miRNA'
)

df_eda['sample_and_data'] = (
    df_eda['sample_acronym'] + df_eda['data_acronym']
)
```

```python
df_eda.columns
```

# Projetos

```python
# Quantidade de casos e arquivos de interesse por projeto
df_eda \
    .groupby('project_id') \
    .agg(
        case_count=pd.NamedAgg(column='case_id', aggfunc='nunique'),
        file_count=pd.NamedAgg(column='file_id', aggfunc='nunique')
    ) \
    .sort_values(by='case_count', ascending=False)
```

```python
# Aumenta o número máximo de linhas no display
pd.set_option('display.max_rows', 500)

# Quantidade de casos e arquivos por tipo de dados em cada projeto
df_eda \
    .groupby(['project_id', 'sample_type', 'data_type']) \
    .agg(
        case_count=pd.NamedAgg(column='case_id', aggfunc='nunique'),
        file_count=pd.NamedAgg(column='file_id', aggfunc='nunique')
    ) \
    .sort_values(by=['project_id', 'sample_type'], ascending=True)
```

```python
df_eda.value_counts(['project_id', 'sample_type', 'data_type'])
```

# Casos

```python
df_test = df_eda[['case_id', 'file_id', 'sample_and_data']].copy()

df_test['TTmRNA'] = np.where(df_test['sample_and_data'] == 'TTmRNA', 1, 0)
df_test['TTmiRNA'] = np.where(df_test['sample_and_data'] == 'TTmiRNA', 1, 0)
df_test['NTmRNA'] = np.where(df_test['sample_and_data'] == 'NTmRNA', 1, 0)
df_test['NTmiRNA'] = np.where(df_test['sample_and_data'] == 'NTmiRNA', 1, 0)

df_test['file_count'] = (
    df_test['TTmRNA'] + df_test['TTmiRNA'] + df_test['NTmRNA'] + df_test['NTmiRNA']
)

# Quantidade de arquivos de interesse por caso
df = df_test \
    .groupby('case_id') \
    .agg(
        file_count=pd.NamedAgg(column='file_count', aggfunc='sum'),
        TTmRNA_count=pd.NamedAgg(column='TTmRNA', aggfunc='sum'),
        TTmiRNA_count=pd.NamedAgg(column='TTmiRNA', aggfunc='sum'),
        NTmRNA_count=pd.NamedAgg(column='NTmRNA', aggfunc='sum'),
        NTmiRNA_count=pd.NamedAgg(column='NTmiRNA', aggfunc='sum')
    ) \
    .sort_values(
        by=['file_count', 'TTmRNA_count', 'TTmiRNA_count', 'NTmRNA_count', 'NTmiRNA_count'],
        ascending=False
    )

df
```

```python
count_abs = df.value_counts(ascending=False)
count_norm = df.value_counts(normalize=True, ascending=False)

df_counts = pd.DataFrame({
    'count': count_abs,
    'count%': round(count_norm * 100, 2)
})

df_counts = df_counts.reset_index()
df_counts
```
